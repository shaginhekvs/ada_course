# Scrape universities' specific data  from QS
#### and save it in unis_qs_main

```python
missing=True
while(missing):
    for i,uni_data in enumerate(unis_qs_main):
        url_base = "https://www.topuniversities.com/{}"
        url_tag=uni_data['url']
        response = requests.request("GET", url_base.format(url_tag))
        soup=BeautifulSoup(response.text,'html.parser')
        n_fac,n_int_fac,n_stu,n_int_stu=0,0,0,0
        try:
            n_fac=int(soup.find('div',class_='total faculty').find('div',class_='number').text.strip().replace(',',''))
        except:
            print('problem with n_fac {} at index {}'.format(i,uni_data['title']))
            missing=True
        try:
            n_int_fac=int(soup.find('div',class_='inter faculty').find('div',class_='number').text.strip().replace(',',''))
        except:
            print('problem finding n_int_fac {} at index {}'.format(i,uni_data['title']))
            missing=True
        try:    
            n_stu=int(soup.find('div',class_='total student').find('div',class_='number').text.strip().replace(',',''))
        except:
            print('problem finding n_stu {} at index {}'.format(i,uni_data['title']))
            missing=True
        try:
            n_int_stu=int(soup.find('div',class_='total inter').find('div',class_='number').text.strip().replace(',',''))
            missing=False
        except:
            print('problem finding n_int_stu {} at index {}'.format(i,uni_data['title']))
            missing=True
        uni_data['n_fac']=n_fac
        uni_data['n_int_fac']=n_int_fac
        uni_data['n_stu']=n_stu
        uni_data['n_int_stu']=n_int_stu
```

# Download names of universities from wikipedia
#### Use wikipedia's knowledge graph to download names of both times and qs universities and use that to later align the 2 dataframes. 
#### qs_wiki_titles and times_wiki_titles has these names as a dictionary where keys are the names of the universities and the values are the wikipedia's page's title. 

```python
for i in range(500):
    try:
        for name in names_qs:
            if (not (name == 'UCL')) and (name not in qs_wiki_titles.keys()):
                qs_wiki_titles[name]=wikipedia.page(wikipedia.search(name)[0]).title

        for name in names_times:
            if name not in times_wiki_titles.keys():
                times_wiki_titles[name]=wikipedia.page(wikipedia.search(name)[0]).title
    except Exception as e:
        pass
        print('couldn\'t load for {}'.format(name))
qs_wiki_titles['UCL']='University College London' #UCL has ambiguous entries so manually add. 
```

  ####  Obtain the 200 top-ranking universities in www.topuniversities.com (ranking 2018). In particular, extract the following fields for each university: name, rank, country and region, number of faculty members (international and total) and number of students (international and total). Some information is not available in the main list and you have to find them in the details page. Store the resulting dataset in a pandas DataFrame and answer the following questions:

###### -Which are the best universities in term of: (a) ratio between faculty members and students, (b) ratio of international students?
###### -Answer the previous question aggregating the data by (c) country and (d) region.

We load the the raw data from www.topuniversities.com using as the index the name of the university according to its wikipedia article (explanation in Exercise 3). We clean up the data dropping duplicate columns. The resulting dataframe contains the following columns:
* country: where the university is
* region: continent where the university is
* n_int_fac: number of international faculty members
* n_int_stu: number of international students
* n_stu: number of students
* n_fac: number of faculty members
* rank_display: number of the university in the rank
* wiki_name: name of the wikipedia article refering to that university

We compute the student-faculty ratio by dividing the total number of students by the number of faculty from each university. The proportion of international students is likewise computed dividing the number of international students by the total number of students. We add each of these values as a column to the dataframe with the names 'ratio_stu_fac' for the student-faculty ratio and 'ratio_int_stu' for the proportion of international students. In addition, we change any column with an infinite value due to not having the number of faculty or studens.


```python
# Load the qs data and clean it up
qs=pd.read_csv('qs_top200_with_wiki.csv',sep='~',encoding = "ISO-8859-1",index_col=9)
qs.drop('Unnamed: 0',1, inplace = True) #Drop duplicate column
qs.drop('title',1, inplace = True) #Drop duplicate column
qs.drop('source',1, inplace = True) #Drop source column

#Add ratios to data frame
qs['ratio_stu_fac']=qs.n_stu/qs.n_fac#Student-faculty ratio
qs.ratio_stu_fac=qs.ratio_stu_fac.replace(np.inf, np.nan)
qs['ratio_int_stu']=qs.n_int_stu/qs.n_stu #International student ratio
qs.ratio_int_stu=qs.ratio_int_stu.replace(np.inf, np.nan)
qs.head()
```

### Which are the best universities in term of: (a) ratio between faculty members and students, (b) ratio of international students?

#### a) Student - faculty ratio

We show the rank obtained ordering universities by student faculty ratio, ordering them in ascending order, i.e. considering better to have less students per faculty member. We drop the universities with NaN, that, thus, could not be ranked with this ratio. According to qs, the best universities in terms of student - faculty ration are __Hokkaido University__,  __Pohang University of Science and Technology__ and __Duke University__. 



```python
#Student-faculty ratio 
qs_stu_fac=qs.ratio_stu_fac.sort_values(ascending=True).dropna()

print('Rank by student-faculty ratio:\n', qs_stu_fac)
```

To study the results we plot the ratios obtained both using a bar chart for the top 50 universities, to show how it tends to grow in the best ranked, and also for all the universities to observe the global trend. 
In addition, we also plot the rank obtained with this criteria by each university vs the rank which they had in www.topuniversities.com, to try to see the correlation between having a few students per faculty member (low student-faculty ratio, high position on the rank) had a great influence in the global rank found in www.topuniversities.com. In the last two plots, instead of putting the names of the universities in the x axis we put the position on the student faculty ranking for readibility reasons.

```python
#Global rank of the universities ordered by the student-faculty ratio
rank=[qs.loc[x].rank_display for x in qs_stu_fac.index]

#Plot the ratios for the top 50
plt.rcParams.update({'font.size': 15}) #Increment font size
plt.figure(figsize=(24,7)) #Bigger plot
qs_stu_fac[:50].plot(kind='bar')
plt.title('Top 50 universities in the student faculty ratio')
plt.ylabel('Number of students per faculty member')


#Plot the ratios
plt.rcParams.update({'font.size': 15}) #Increment font size
plt.figure(figsize=(24,7)) #Bigger plot
plt.plot(range(1,len(rank)+1), qs_stu_fac)
plt.title('Top 50 universities in the student faculty ratio')
plt.title('Student faculty ratio')
plt.ylabel('Number of students per faculty member')
plt.xlabel('Rank with the student faculty ratio')

#Plot the rank according to the student-faculty ratio vs the global rank found in www.topuniversities.com
plt.figure() 
plt.figure(figsize=(24,7)) #Bigger plot
plt.stem(range(1,len(rank)+1), rank)
plt.title('Rank according to the student-faculty ratio vs the global rank found in www.topuniversities.com')
plt.xlabel('Rank with student-faculty ratio')
plt.ylabel('Global rank')
```

* **Discussion** 

The first plot shows that the growth in the number of students per faculty member is slow, with a small slope. Despite this, the ratio in the third university doubles the one found in the first. It is worth noting that the best ranked universities seem to be from US, with 3 in the top 5 and 6 in the top 10.

The second plot allows us to observe the general trend, seeing that the number of students per faculty grows slowly, with a small slope and with less students per faculty member until we reach around the 180th position on the ranking, where there is an exponential growth.

In the third plot we can observe that there is not a clear correlation between having a low student-faculty ratio and having a high rank in www.topuniversities.com, even though there seems to be more correlation in the first positions. However, this can be related to the fact that the differences in this ratio are very small, as can be observed in the first plot. Thus, being in a low position on the student-faculty ratio can still imply a very low number of students per faculty member. 

#### b) International student ratio
In this case, we order the ranking in descending order, as we consider better having a high proportion of international students. Following the same procedure as with the student-faculty ratio, we drop the universities which obtained NaN due to missing the number of international students or the number of students.

Again, to analyse the results we use three plots plots. The first one shows the top 50 universities in the ranking. The second one shows what is the proportion of international students in descending order in the ranking. The third one compares the ranking obtained with this ratio in the x axis and the one obtained in www.topuniversities.com in the y axis.

```python
#Ratio of international students
qs_int_stu=qs.sort_values(by='ratio_int_stu',ascending=False).dropna()
print('Rank by ratio of international students:\n', qs_int_stu.ratio_int_stu)

#Plot the ratios for the top 50

plt.figure(figsize=(24,7)) #Bigger plot
plt.rcParams.update({'font.size': 15}) #Increment font size
qs_int_stu.ratio_int_stu[:50].plot(kind='bar')
plt.title('Top 50 universities in the international students ratio')
plt.ylabel('Proportion of international students')


#Plot the ratios
plt.figure(figsize=(24,7)) #Bigger plot
plt.plot(range(1,len(qs_int_stu)+1), qs_int_stu.ratio_int_stu)
plt.title('International students ratio')
plt.ylabel('Proportion of international students')
plt.xlabel('Rank with the international students ratio')

#Global rank of the universities ordered by the ratio of international students
rank=[qs.loc[x].rank_display for x in qs_int_stu.index]

#Plot the rank according to the ratio of international students vs the global rank found in topuniversities.com
plt.figure(figsize=(24,7)) #Bigger plot
plt.stem(range(1,len(rank)+1), rank)
plt.title('Rank according to the international students ratio vs the global rank found in www.topuniversities.com')
plt.xlabel('Rank with the international students ratio')
plt.ylabel('Global rank')
```

* ** Discussion **

The first plot shows a quick descense of the proportion of the international students until it stabilizes with a smooth slow descent when it reaches 40% of international students, being the differences in the rank very small for the following universitites. The top 4 are the same found with the student-faculty ratio, which shows a correlation between the features. In this case, the UK universities tend to be better rated with 3 universities in the top 5 and 5 in the top 10.

In the second plot we can observe the opposite behaviour to the one found with the student-faculty ratio, the results show an exponentional shape, descending with a steep decrease at first and then having a much slower descend. However, the differences in values are much greater than with the student-faculty ratio, ranging from 70% to 10%.

In the third plot we can observe, as with the student-faculty ratio, that there is not a clear correlation between obtaining a high rank in the international student ratio and a high rank in www.topuniversities.com.

### Answer the previous question aggregating the data by (c) country and (d) region.

#### Student - faculty ratio

To compare the results in c) and d) we first aggregate the student-faculty ratio by country and by region. In order to do so, we aggregate the results with __groupby()__. For the plots, we use compute the mean of each of the groups, but we also show in tables other statistics with __describe()__, to understand better the distribution of the data. As before, we drop the cases where there is no information about this ratio and sort the data by increasing student faculty ratio.

#### c) Aggregate by country

```python
#Aggregate the student-faculty ratio by country
agg_country_stu_fac_qs= qs.dropna().groupby(by='country').ratio_stu_fac

plt.rcParams.update({'font.size': 10}) #Increment font size
plt.figure(figsize=(10,5)) #Bigger plot
agg_country_stu_fac_qs.mean().sort_values(ascending=True).dropna().plot(kind='bar')
plt.title('Student-faculty ratio by country')
plt.xlabel('Country')
plt.ylabel('Mean number of students per faculty member')

print("Student-faculty ratio by country:\n")
agg_country_stu_fac_qs.describe().sort_values(by = 'mean' , ascending=True)
```

* **Discussion**

In the table can be observed that the ratio in countries seems to be very stable, with low standard deviations and with close values between mean and median. This indicates that the mean is probably a good measure of the student-faculty ratio.

In the plot can be observed that as happened when observing the student-faculty ratio by university it has a small slope, where values between countries are very similar, until in the last countries it has a steep increase showing a exponential behaviour.  Surprisingly, given the position of its universities in the individual ranking, United States is not among the first ones in the ratio.

#### d) Aggregate by region



```python
#Aggregate the student-faculty ratio by region
agg_region_stu_fac_qs=qs.groupby(by='region').ratio_stu_fac

plt.figure(figsize=(10,3)) #Bigger plot
agg_region_stu_fac_qs.mean().sort_values(ascending=True).dropna().plot(kind='bar')
plt.title('Student-faculty ratio by region')
plt.xlabel('Region')
plt.ylabel('Proportion')

print("Student-faculty ratio by region: \n") 
agg_region_stu_fac_qs.describe().dropna().sort_values(by = 'mean', ascending=True)
```

* **Discussion**

In this case, the table shows that the values are not so stable as in the previous case. In Latin America there is the greatest variability of the ratio across universities.

The plot shows the same shape as when aggregating by country with very similar, but increasing values between Asia and Europe and North America, which have very similar values, and a steep increase in Oceania and Latin America.

#### Obtain the 200 top-ranking universities in www.timeshighereducation.com (ranking 2018). Repeat the analysis of the previous point and discuss briefly what you observed.

We load the the raw data from www.timeshighereducation.com following the same procedure as with www.topuniversities.com, that is we use the name of the wikipedia of the university as index and clean up the data dropping duplicate columns. To be able to compare them better, we rename the columns from the resulting dataframe so they have the same names as www.topuniversities.com dataframe. # Exercise 3
#### Merge the two DataFrames created in questions 1 and 2 using university names. Match universities' names as well as you can, and explain your strategy. Keep track of the original position in both rankings.

To merge the two DataFrames we encountered the problem of the same university having different names in each of the rankings, for example, having the acronym in one of them and not in the other,  or directly being in english in one and not in the other. In order to solve this problem to be able to match universities in both rankings by their names, we added a column when scraping the data to each of the dataframes, which contains the title of the wikipedia article refering to that university. 

As we only want the universities present in the two rankings, we do an __inner__ merge on the wikipedia name, which we use as index, adding as a suffix \__qs_ to www.topuniversities.com columns and \__the_ to www.timeshigheruniversities.com columns. 

```python
#Inner merge
merge_df=qs.join(the, how='inner', lsuffix='_qs', rsuffix='_the', sort=True)
merge_df.head()
```

#### Comparison of the position of the same university in the rankings

To compare the rankings we plot in the first place the ranks of the top 50 universities found in www.topuniversities.com in front of the one found in www.timeshigheruniversity.com, sorting by the first one, to observe whether universities get similar rankings in both universites. Universities that have the same rank are ordered according to their names one after the other. In addition, we also compute the difference in the ranks and study how much they usually diverge with _describe()_.

```python
#Ranks from both rankings sorted by qs
rank_sorted_by_qs=merge_df[['rank_display_qs', 'rank_display_the']].sort_values(by='rank_display_qs', ascending=True)

#Plot
plt.figure(figsize=(20,10)) #Bigger plot
plt.stem(range(0,50), rank_sorted_by_qs[:50].rank_display_the) #Range due to duplicate ranks
plt.title('Relation between position in both rankings')
plt.xlabel('Rank in www.topuniversities.com')
plt.ylabel('Rank in www.timeshigheruniversities.com')

plt.xticks(range(0,50), range(1,51))


#Difference between ranks
print('Difference between the ranks in www.topuniversities.com and www.timeshigheruniversities.com')
np.abs(merge_df['rank_display_qs']-merge_df['rank_display_the']).describe()
```

* **Discussion**

In the plot can be observed that for the top universities of www.topuniversities.com, the rank is also high in www.timeshigheruniversities,com, even though they are not in the same order. However, as the ranks grows there seems to be a bigger divergence between the two rankings. In total, we can see in the table that the mean divergence is around 32 positions, which indicates that the rankings must have very different criteria.

# Exercise 5
#### Can you find the best university taking in consideration both rankings? Explain your approach.

We selected __Massachusetts Institute of Technology__ as being the best university in the world, due to the fact that it has the highest rank based on the harmonic mean of both ranking websites. 

The reason that we decided to select our statistic to be the harmonic mean is because it penalises low values (the university has to be very good in both rankings). Thus, we believed it is the best way to capture most information out of both rankings and to account for different criteria. We also considered the option of performing a regression on the data to simulate the ranking process but we came to the conclusion that there are too many parameters that are missing in the analysis for it to be meaningful. Thus, we chose a more holistic approach. 

```python
top5 = merge_df_insights.sort_values(by ='rank_display_the')[0:5]
top5 = top5[["rank_display_qs" ,"rank_display_the"] ]
top5['rank_harmon'] = np.sqrt(top5["rank_display_qs"] * top5["rank_display_the"] )
top5.sort_values(by='rank_harmon')
```
