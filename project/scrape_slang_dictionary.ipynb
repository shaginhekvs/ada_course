```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pickle
import string
```

```python
def get_slang():    
    '''
    Get a slang dictionary through web scraping.
    
    INPUT:
    OUTPUT: 
        dictionary
    
    '''
    
    #Dictionary slang-word
    slang={}
    #Base URL 
    base_URL='https://www.noslang.com/dictionary/'

    #Webpage sorted by numbers and then letters of the alphabet
    for letter in '1'+string.ascii_lowercase:
        #Request from current letter
        r = requests.get(base_URL+letter+'/')
        #Parser
        soup=BeautifulSoup(r.text,'html.parser')
        
        #Add every word found to the dictionary
        for word in soup.find_all('div',class_='dictionary-word'):
            #Add the words to the dictionary. For the slang, delete : and whitespace. For the words, replace ** in curses
            slang[word.find('dt').text.replace(':','').strip()]=word.find('dd').text.replace('f**k','fuck'). \
                                                                                      replace('a**', 'ass'). \
                                                                                      replace('h**e','hole'). \
                                                                                      replace('s**t', 'shit'). \
                                                                                      replace('b***t', 'bitc'). \
                                                                                      replace('b*****d', 'bastard'). \
                                                                                      replace('c**k', 'cock'). \
                                                                                      replace('d**k', 'shit'). \
                                                                                      replace('d****e', 'douche'). \
                                                                                      replace('d**n', 'damn'). \
                                                                                      replace('p***y', 'pussy'). \
                                                                                      replace('n***a', 'nigga'). \
                                                                                      replace('n****r', 'nigger') 
    return slang
```

```python
slang=get_slang()
```
