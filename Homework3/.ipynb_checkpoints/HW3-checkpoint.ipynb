Unemployment rates are major economic metrics and a matter of concern for governments around the world. Though its definition may seem straightforward at first glance (usually defined as the number of unemployed people divided by the active population), it can be tricky to define consistently. For example, one must define what exactly unemployed means : looking for a job ? Having declared their unemployment ? Currently without a job ? Should students or recent graduates be included ? We could also wonder what the active population is : everyone in an age category (e.g. 16-64) ? Anyone interested by finding a job ? Though these questions may seem subtle, they can have a large impact on the interpretation of the results : 3% unemployment doesn't mean much if we don't know who is included in this percentage.

In this homework you will be dealing with two different datasets from the statistics offices of the European commission (eurostat) and the Swiss Confederation (amstat). They provide a variety of datasets with plenty of information on many different statistics and demographics at their respective scales. Unfortunately, as is often the case is data analysis, these websites are not always straightforward to navigate. They may include a lot of obscure categories, not always be translated into your native language, have strange link structures, … Navigating this complexity is part of a data scientists' job : you will have to use a few tricks to get the right data for this homework.

We use a Total unemployment rate table (teilm20) found under Tables by themes -> Population and social conditions -> Labour market (t_labour) -> Employment and unemployment (Labour Force Survey) Unemployment -> LFS main indicators -> LFS adjusted series (t_une) -> Harmonised unemployment rate by sex (teilm020) (http://ec.europa.eu/eurostat/data/database?node_code=teilm020).

According to its description this table definition of unemployement rate is:

>The unemployment rate represents unemployed persons as a percentage of the labour force based on International Labour Office (ILO) definition. The labour force is the total number of people employed and unemployed. Unemployed persons comprise persons aged 15 to 74 who:
- are without work during the reference week; 
- are available to start work within the next two weeks; 
- and have been actively seeking work in the past four weeks or had already found a job to start within the next three months. 

It holds the unemployment rate over the last 12 months, measured as the percentage of the active population. The data is also categorized by sex (female, male or total).

First, we load the table to a pandas dataframe named 'total_unemployment_EU_raw'., separating the columns by the commas and also by the tabulations using _read_\__csv()_ .We clean up the dataframe, first dividing the country and sex columns in two and afterwards deleting whitespace from the column names and from the countries column.

```python
#Read data: separators are commas and tabulators
data_pth='Data/une_rt_m.xls'
total_unemployment_EU=pd.read_excel(data_pth, indexcol=0, skiprows=np.arange(10))
total_unemployment_EU.head()
```

```python
#Divide GEO and SEX columns by -
total_unemployment_EU['id'],total_unemployment_EU['GEO']=total_unemployment_EU['GEO'].str.split('-', 1).str
total_unemployment_EU['sex_id'],total_unemployment_EU['SEX']=total_unemployment_EU['SEX'].str.split(' - ', 1).str

#Rename germany columns
total_unemployment_EU.loc[30:32,'GEO']='Germany'
#Remove whitespace from column titles
total_unemployment_EU=total_unemployment_EU.rename(columns=lambda x: x.strip())
total_unemployment_EU['GEO']=total_unemployment_EU['GEO'].str.strip()

total_unemployment_EU.head()
```

Even though the most recent data is from September, it has missing data in some countries, so we use the data from July, which is the closest with data from all countries.

```python
any(total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')]['2017M09'].isnull())
```

```python
any(total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')]['2017M08'].isnull())
```

```python
any(total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')]['2017M07'].isnull())
```

We then load the json data:

```python
#Load json
EU_geo_path =  os.path.join("topojson","europe.topojson.json")
geo_json_data_EU = json.load(open(EU_geo_path))
```

Then, for the map, we find which countries in the json file are not present in our database, so we can plot them in a different colour, and add them to a pandas DataFrame called 'no_countries_df'

```python

#Select GEO and unemployment rate for September for total data
data_plot=total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')][['GEO','2017M07']]

#We compute which countries from json file are not available 
countries_not_available=[]
list_countries_avail=list(total_unemployment_EU.GEO) #Countries in the database
#Iterate through each country in json file
for info in geo_json_data_EU['objects']['europe']['geometries']:
    country=info['properties']['NAME']
    #Country not in database
    if  country not in list_countries_avail:
        countries_not_available.append(country)

#Dataframe with names of not available countries
no_countries_df=pd.DataFrame({'GEO':countries_not_available,'value':np.ones(len(countries_not_available))})       

```

We create our own colormaps. For the values present, we use a divergin colormap which ranges from low values of the unemployment rate in green to high values in red, with intermediate values in yellow. We plot the missing data in blue. 

To assign this colors we create a function, called 'my_color_function' which returns the colour assigned to that country depending on if it's missing or not and on the unemployment rate. This function receives:
* feature: country in the json file
* color_map: colormap for not missing countries
* data_col: column of dataframe with countries ids
    

```python

#Create colormap from green (minimum rate) to red (maximum unemployment rate)
linear_cm = cm.LinearColormap(
    [ 'green', 'yellow', 'red'],
    vmin=0, vmax=data_plot['2017M07'].max()
)

#Missing data in blue
missing = cm.LinearColormap(['blue','blue'])
missing.caption='No data available'
linear_cm=linear_cm.to_step(
    n=7)
linear_cm.caption='Unemployment rates'

#Assign the colors
def my_color_function(feature,color_map,data_col):
    """Maps low values to green and high values to red."""
    if feature['properties']['NAME'] in countries_not_available:
        return '#0000FF'
    else:
        return color_map(data_col.loc[data_col.GEO==feature['properties']['NAME']]['2017M07'].iloc[0])
```

Finally, we map the unemployment rate for the month of July. For this purpose, we use a layered map, mapwith one layer showing the unemployment rate among women, other among men and the last one in total for both women and men. 

For comparison, we use the same the same threshold and colorbar for the three cases with women, men and in total. 

```python

#Create map
central_location = [50.519962, 6.633597]
map_EU= folium.Map(central_location,tiles='cartodbpositron', zoom_start=3)

#Data of men and women
data_T=total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')]
data_plot=total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'T')][['GEO','2017M07']]


#Clorophet
map_T=folium.TopoJson(geo_json_data_EU,'objects.europe',
                name='Total Unemployment',
                style_function= lambda feature: {
                'fillColor': my_color_function(feature,linear_cm,data_plot),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })


#Data of women 
data_F=total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'F')][['GEO','2017M07']]


map_m=folium.TopoJson(geo_json_data_EU,'objects.europe',
                name='Women\'s Unemployment',
                style_function= lambda feature: {
                'fillColor': my_color_function(feature,linear_cm,data_F),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })

#Data of men 
data_M=total_unemployment_EU.loc[(total_unemployment_EU['sex_id'] == 'M')][['GEO','2017M07']]

map_f=folium.TopoJson(geo_json_data_EU,'objects.europe',
                name='Men\'s Unemployment',
                style_function= lambda feature: {
                'fillColor': my_color_function(feature,linear_cm,data_F),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })
map_EU.add_child(map_T)
map_EU.add_child(map_m)
map_EU.add_child(map_f)
map_EU.add_child(missing)
map_EU.add_child(linear_cm)

folium.LayerControl().add_to(map_EU)
map_EU
```



In the map we can observe with redder colours the countries with worse unemployement rate and with greener the lower. Countries completely in blue are those with missing data, like Switzerland, so they cannot be compared.

A tendency can be observed where the countries in the south of Europe, in the mediterranean area, tend to have a higher unemployment rate, with the highest being Greece with 21%, followed by Spain with a 18.7%. On the other hand, Germany and Austria are the countries with lowest unemployment rates. 

Between the countries, the tendency is the same if we only compare data of men and women. However, it is worth noting that the unemployment rate is higher among women in the countries with higher unemployment rates, such as Spain, Italy and Croatia, in comparison to the one found with men. 

To study better this possibility, we also show the difference between the unemployment rate found for women and men in a new map. First, we create a new dataframe with this difference and afterwards, plot it in a map. We use a colour scale, ranging from the minimum to maximum difference, divided by the quantiles. Countries are in red tones when there is less unemployment for women and in green tones when the unemployment rates are smaller in men.

```python
#Difference between women and men
diffEU_gender =pd.DataFrame(data = {'GEO': data_F.GEO.values, '2017M07':data_M['2017M07'].values-data_F['2017M07'].values})

diff_cm=cm.LinearColormap(['green','yellow','red']).to_step(25,data=diffEU_gender['2017M07'].sort_values().dropna().values,method='quantiles',\
               round_method='int').scale(min(diffEU_gender['2017M07']), vmax=max(diffEU_gender['2017M07']))


#Map
map_EU_diff= folium.Map(central_location,tiles='cartodbpositron', zoom_start=3)

map_diff=folium.TopoJson(geo_json_data_EU,'objects.europe',
                name='Unemployment difference',
                style_function= lambda feature: {
                'fillColor': my_color_function(feature,diff_cm,diffEU_gender),
                'color': 'black',
                'fill_opacity':10,
                'weight': 2,
                'dashArray': '5, 5' 
               })
map_EU_diff.add_child(map_diff)
map_EU_diff.add_child(missing)
map_EU_diff.add_child(diff_cm)


map_EU_diff
```

In the map we can observe that, as observed before, in the south of Europe the unemployment rates tend to be bigger amongst women. On the other hand, in the center and north of Europe, rates tend to very similar between women and men, with slightly better values for women.

It should be noted, that while one can find countries with 8% more of unemployment rate in women, only 2% higher can be found where men are more unemployed.

From the amstat webpage we have downloaded an excel file which contains the unemployment rate, the number of unemployed people, the number of people looking for a job and the number of people looking for a job who already have one for the month of __September__. We chose this month for both exercises, for it is the most recent and also does not have the summer trends were there tend to be more fluctuations due to holidays. We load this data in a dataframe, deleting the empty column and renaming the column with the name of the cantons. Additionally, we also delete the row with the totals of each column.

The resulting dataframe 'total_unemployment_CH' has the following columns: 
* Canton: name of the canton
* Taux de chômage: unemployment rate
* Chômeurs inscrits: number of unemployed people registered
* Demandeurs d'emploi: number of people looking for a job
* Demandeurs d'emploi non chômeurs: number of people looking for a job currently employed

```python
#Read data
data_pth='Data/Ex2.xlsx'
total_unemployment_CH_raw=pd.read_excel(data_pth, indexcol=0, skiprows=[0,1,2],usecols=np.arange(6))


#Rename columns
total_unemployment_CH=total_unemployment_CH_raw.drop('Mesures',axis=1)
total_unemployment_CH.rename(columns={'Unnamed: 0':'Canton'}, inplace=True)

#Drop total row
total_unemployment_CH=total_unemployment_CH.drop(len(total_unemployment_CH)-1,axis=0)
total_unemployment_CH.sort_values(by='Taux de chômage').head()
```

Then, we load the json data.

```python
#Load json data
CH_json_path = os.path.join("topojson","ch-cantons.topojson.json")
geo_json_data_CH = json.load(open(CH_json_path))
```

Afterwards, we compare the names in the json file and the dataframe to see if they match. We print the name in each of the files. As can be observed, there is not an exact match between the name in the json file and in the dataframe, due to special characters like tildes, and because in the json file the names in some cases appear both in french and german. Therefore, we need to match the names in one file to the other one. However, we can see that each of the rows, even though they have different names refer to the same canton, so we can simply add a column with the id's to dataframe, which we call 'id'.


```python
#Print id, name in json and name in dataframe
print('\033[1m' ) #Bold
print('{:>1}  {:>30}  {:>30}'.format('ID', 'JSON NAME','DATAFRAME NAME'))
print('___________________________________________________________________')
print('\033[0m') #End bold

#Loop through each row in the json file
for (i, row) in enumerate(geo_json_data_CH['objects']['cantons']['geometries']):
    print('{:>1}  {:>30}  {:>30}'.format(row['id'], row['properties']['name'],total_unemployment_CH['Canton'][i]))

#Add cantons id's
total_unemployment_CH['id']=[row['id'] for row in geo_json_data_CH['objects']['cantons']['geometries']] 
```

Before plotting the map, in order to compare both cases better, we compute the unemployement rate without considering also the people who are already employed but are looking for a job. To do so,  we first get the total amount of active population as the total number of people looking for employment divided by the unemployement rate found in amstat. 

```python
#Total active poblation per Canton
total=total_unemployment_CH['Demandeurs d\'emploi']*100/total_unemployment_CH['Taux de chômage']
```

With this value, we compute the unemployment rate considering only the people who do not have a job and add it to our data frame under the name 'rate_unemployed'.

```python
#Unemployment rate of people without jobs
total_unemployment_CH['rate_unemployed']=total_unemployment_CH['Chômeurs inscrits']*100/total
total_unemployment_CH.sort_values(by='rate_unemployed',ascending=False).head()
```

```python
#Assign the colors
def my_color_function_CH(feature,color_map,data_col):
    """Maps low values to green and high values to red."""
    if feature['geometries']['NAME'] in countries_not_available:
        return '#0000FF'
    else:
        return color_map(data_col.loc[data_col.GEO==feature['geometries']['id']]['2017M07'].iloc[0])
```

Then, we can finally plot the map with the unemployment rate per canton. As before, we use a layered map, showing in red worse rates and in green the best rates.

```python
#Create map
m_switzerland = folium.Map([46.799962, 8.1233597], tiles='cartodbpositron', zoom_start=6.7)

#Colormap from green to red divided by quartiles
CH_cm=cm.LinearColormap(['green','yellow','red']).to_step(25,data=total_unemployment_CH['Taux de chômage'].sort_values().dropna().values,method='quantiles',\
               round_method='int').scale(min(total_unemployment_CH['Taux de chômage']), vmax=max(total_unemployment_CH['Taux de chômage']))

#Cloropleths
map_with_employed_CH=folium.TopoJson(geo_json_data_CH,'objects.cantons',
                name='With employed',
                style_function= lambda feature: {
                    
                'fillColor': CH_cm(total_unemployment_CH.loc[total_unemployment_CH.id==feature['id']]\
                                       ['Taux de chômage'].iloc[0]),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })
map_unemployed_CH=folium.TopoJson(geo_json_data_CH,'objects.cantons',
                name='Only unemployed',
                style_function= lambda feature: {
                    
                'fillColor': CH_cm(total_unemployment_CH.loc[total_unemployment_CH.id==feature['id']]\
                                       ['rate_unemployed'].iloc[0]),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })



m_switzerland.add_child(map_with_employed_CH)
m_switzerland.add_child(map_unemployed_CH)
m_switzerland.add_child(CH_cm)


folium.LayerControl().add_to(m_switzerland)
m_switzerland
```

Counting all people looking for a job, it can be observed that the unemployement rates in Switzerland are very low. The French and Italian speaking regions tend to have a higher unemployment rate, with the lowest being in the german speaking part.

If we only considered unemployed people looking for a job, we find that some cantons in the german part are missing data, shown as before in deep red. The unemployement rates in all the other cantons, are, of course, lower, but we can still find the same trend with higher rates in the french speaking part. It is worth mentioning that in the cantons of Tessin and Zurich, which we amongst the ones with higher rates in the first case, have a much smaller rate when only considering unemployed people.

Next, to compare how the unemployment rate changes when we consider people who already have a job, we plot in a bar chart the rates with and without people who already have a job. Additionally, we show a map with the difference of the two rates, using again equally spaced bins.

As before, it is clear that when you only considered unemployed people the rates have a great descent. Contrary to what we noticed in the first map, with these figures we can notice this descent is not specially steep in  Zurich, although it is in Tessin. Nevertheless, is in small cantons like Uri, Obwald and Grissons were we find the biggest difference, with rates being reduced by half.

```python
#Bar chart for comparing the rates with employed or not people
ax=total_unemployment_CH.set_index('Canton')[['Taux de chômage','rate_unemployed']].plot.bar(figsize=(15,7))
ax.legend(["With employ", "Only unemployed"])
ax.set_ylabel("Rate of unemployement (%)") 
```

```python
#Difference between women and men
diffCH_employ =pd.DataFrame(data = {'id': total_unemployment_CH.id.values, 
                                    'diff':total_unemployment_CH['Taux de chômage'].values-\
                                    total_unemployment_CH['rate_unemployed'].values})

#Colormap
CH_cm_diff=cm.LinearColormap(['green','yellow','red']).to_step(25,data=diffCH_employ['diff'].sort_values().dropna().values, \
               method='quantiles').scale(min(diffCH_employ['diff']), vmax=max(diffCH_employ['diff']))

#Map
map_CH_diff= folium.Map([46.799962, 8.1233597],tiles='cartodbpositron', zoom_start=8)
clorophet_diff_CH=folium.TopoJson(geo_json_data_CH,'objects.cantons',
                name='Difference',
                style_function= lambda feature: {
                    
                'fillColor': CH_cm_diff(diffCH_employ.loc[diffCH_employ.id==feature['id']]\
                                       ['diff'].iloc[0]),
                'color': 'black',
                'fill_opacity':0.7,
                'weight': 2,
                'dashArray': '5, 5' 
               })


map_CH_diff.add_child(clorophet_diff_CH)
map_CH_diff.add_child(CH_cm_diff)
map_CH_diff
```

First we get the data with the same types of columns as in exercise 2, but this time differentiating among swiss(__Suisses__) and non-swiss people (__Etrangers__), to inspect whether there is a consistent difference between the 2 categories. 

```python
# read the data
swiss_foreign_unempl = pd.read_excel('Data/ex3_chômage.xlsx', skiprows=[0,1,2], indexcol=0, usecols=[0,1, 3,5,6,7] )
swiss_foreign_unempl.columns = ['Canton', 'Nationality', 'Unempl Rate', 'Registered Unempl', 'Job seekers', 'Empl job seekers']
swiss_foreign_unempl.head()
```

The formula for calculating the unemplowment rate in this dataset, is the number of people actively looking for a job (__job seekers__) over the active population times 100. However, we stick to the definition that seems more natural to us, that is the number of people actively looking for a job that are currently __unemployed__, over the active population times 100. This is expressed in the column entitled __Unempl Rate jobless__.

```python
active_pop = swiss_foreign_unempl['Job seekers'] * 100 / swiss_foreign_unempl['Unempl Rate']
active_pop
swiss_foreign_unempl['Unempl Rate jobless'] = (swiss_foreign_unempl['Job seekers'] - swiss_foreign_unempl['Empl job seekers'])\
                                                /active_pop * 100
# check that registered unemployers is indeed equivalent to the unemployed job seekers
sum(swiss_foreign_unempl['Job seekers']- swiss_foreign_unempl['Empl job seekers'] != swiss_foreign_unempl['Registered Unempl'])
```

### Age groups

###### The raw data is downloaded from the amstat website and total active population is calculated from the previous data, using the total active population, the rate is then calculated for each category here. It was verified if rate across all the categories are summed together, the total rate is same as 'Taux de chômage'

```python
unemployment_CH_raw=pd.read_excel('./Data/full_data_age_nationality.xlsx',indexcol=0, skiprows=[0,1,2]) # load full data
```

```python
'''
load the data in this cell and clean it to keep only interesting columns 
'''

unemployment_CH_raw=unemployment_CH_raw.rename(columns={'Unnamed: 0':'Canton','Unnamed: 1':'Nationality',\
                                    'Unnamed: 2':'Age Code','Unnamed: 3':'Age groups','Unnamed: 4':'Type Code'})

unemployment_CH=unemployment_CH_raw.loc[:,['Canton','Nationality','Age Code', 'Age groups', 'Type Code','Chômeurs inscrits.8', "Demandeurs d'emploi.8",
       "Demandeurs d'emploi non chômeurs.8"]]
unemployment_CH=unemployment_CH.loc[unemployment_CH['Type Code']=='Total'] # filter by Total across all professions

```

```python
'''
calculate the total active population using 'Taux de chômage' and use this population to calculate rate of employment for each 
group/ nationality.

'''
total_unemployment_CH['total active pop']=total_unemployment_CH['Chômeurs inscrits']*100/total_unemployment_CH['Taux de chômage']
active_pop_df=total_unemployment_CH.loc[:,['Canton','total active pop']].copy()
unemployment_CH=unemployment_CH.merge(active_pop_df,on='Canton')
unemployment_CH['rate of unemployment']=unemployment_CH['Chômeurs inscrits.8']*100/unemployment_CH['total active pop']
```

```python
'''
drop columns not neede anymore after rate has been calculated.
'''
rate_unemploy_age_nation_df=unemployment_CH.drop(['Chômeurs inscrits.8',"Demandeurs d'emploi.8","Demandeurs d'emploi non chômeurs.8",'total active pop','Type Code','Age Code'],axis=1)
```

```python
sns.set(font_scale=5)
sns.set_style("whitegrid")

plot=sns.factorplot(x='Canton',y='rate of unemployment',hue='Age groups',col='Nationality',\
              data=rate_unemploy_age_nation_df,kind='bar',col_wrap=2,sharey=True,size=30)
plot.set_xticklabels(rotation=80,size=40)
plot.set_yticklabels(size=50)
```
